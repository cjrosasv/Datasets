**Import of dataset**




```
insurance_data = pd.read_csv("https://github.com/dphi-official/Datasets/blob/master/travel_insurance/Training_set_label.csv?raw=true")
```

# **Data Understanding**

```
#QUIZ 3.1 - Fill the missing values in "Gender" column with "Not specified"
## Let's check the values in this column

insurance_data['Gender'].value_counts(dropna=False)
```

```
insurance_data['Gender'] = insurance_data['Gender'].fillna("Not specified") #we fill the null values with the expression "Not specified"
insurance_data['Gender'].value_counts(dropna=False) #verification of the results
```

```
#QUIZ 3.2 Drop "Distribution Channel, Destination & Agency Type" columns

list_columns = ['Distribution Channel', 'Destination', 'Agency Type'] #we create a list including the columns to be deleted
insurance_data.drop(list_columns, axis=1, inplace=True) #drop columns
insurance_data.info() #verification of the changes
```

```
#QUIZ 3.3. Perform one hot encoding for 'Agency', 'Gender', 'Product Name' using sklearn 

from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(categories='auto')
feature_arr = ohe.fit_transform(insurance_data[['Agency']]).toarray()
feature_labels = ohe.categories_
feature_labels = np.array(feature_labels).ravel()

features_agency = pd.DataFrame(feature_arr, columns=feature_labels)
features_agency
```

```
feature_arr = ohe.fit_transform(insurance_data[['Gender']]).toarray()
feature_labels = ohe.categories_
feature_labels = np.array(feature_labels).ravel()

features_gender = pd.DataFrame(feature_arr, columns=feature_labels)
features_gender
```

```
feature_arr = ohe.fit_transform(insurance_data[['Product Name']]).toarray()
feature_labels = ohe.categories_
feature_labels = np.array(feature_labels).ravel()

features_product = pd.DataFrame(feature_arr, columns=feature_labels)
features_product
```

```
#join datasets

insurance_data = insurance_data.merge(features_agency, right_index=True, left_index=True)
insurance_data
```

```
insurance_data = insurance_data.join(features_gender)
insurance_data = insurance_data.join(features_product)
insurance_data
```
